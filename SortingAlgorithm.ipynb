```python
from timeit import default_timer as timer
```

Quick sort

```python
def quick_sort(a : list, l : int, r : int) :
    pivot = a[(l + r) // 2]
    i, j = l, r
    while i < j :
        while a[i] < pivot :
            i += 1
        while a[j] > pivot :
            j -= 1
        if i <= j :
            a[i], a[j] = a[j], a[i]
            i += 1
            j -= 1
    if i < r :
        quick_sort(a, i, r)
    if l < j :
        quick_sort(a, l, j)
```

```python
for i in range(1, 11) :

    name = f"{i}.txt"
    with open(name, 'r') as f :
        n = int(f.readline())
        a = list(map(float, f.readline().split()))
    start_time = timer()
    quick_sort(a, 0, n - 1)
    end_time = timer()
    for j in range(n) :
        assert a[j] <= a[j + 1] if j + 1 < n else True
    print(f"Time taken to sort {name}: {end_time - start_time:.6f} seconds")
```

Heap sort

```python
def heapify(a, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and a[l] > a[largest]:
        largest = l

    if r < n and a[r] > a[largest]:
        largest = r

    if largest != i:
        a[i], a[largest] = a[largest], a[i]
        heapify(a, n, largest)

def heap_sort(a):
    n = len(a)

    for i in range(n // 2 - 1, -1, -1):
        heapify(a, n, i)

    for i in range(n - 1, 0, -1):
        a[i], a[0] = a[0], a[i]
        heapify(a, i, 0)
```

```python
for i in range(1, 11) :

    name = f"{i}.txt"
    with open(name, 'r') as f :
        n = int(f.readline())
        a = list(map(float, f.readline().split()))
    start_time = timer()
    heap_sort(a, 0, n - 1)
    end_time = timer()
    for j in range(n) :
        assert a[j] <= a[j + 1] if j + 1 < n else True
    print(f"Time taken to sort {name}: {end_time - start_time:.6f} seconds")
```

Merge sort

```python
def merge_sort(a):
    if len(a) > 1:
        mid = len(a) // 2
        L = a[:mid]
        R = a[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                a[k] = L[i]
                i += 1
            else:
                a[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            a[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            a[k] = R[j]
            j += 1
            k += 1
```
